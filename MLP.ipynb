```python
import numpy as np
```

```python
# implementation of a multilayer perceptron (MLP) with one hidden layer
# the main difference is that MLP requires a non-linear  differentiable activation function 
# the possibilities are: sigmoid, tanh, relu (for this implementation)

class MLP:
    def __init__(self, n_inputs, n_hidden, n_outputs, learning_rate=0.01, n_iters=1000, activation_function='relu', early_stopping=False):
        self.n_inputs = n_inputs
        self.n_hidden = n_hidden
        self.n_outputs = n_outputs
        self.learning_rate = learning_rate
        self.n_iters = n_iters
        self.activation_function = self.set_activation_function(activation_function) # set the activation function based on the provided argument
        self.early_stopping = early_stopping 
        
        # initialize weights and biases for hidden layer and output layer
        self.weights_input_hidden = np.random.rand(self.n_inputs, self.n_hidden) * 0.01 # small random weights for input to hidden layer
        self.bias_hidden = np.zeros((1, self.n_hidden)) # zero biases for hidden layer
        self.weights_hidden_output = np.random.rand(self.n_hidden, self.n_outputs) * 0.01 # small random weights for hidden to output layer
        self.bias_output = np.zeros((1, self.n_outputs)) # zero biases for output layer

    def set_activation_function(self, activation_function):
        if activation_function == 'sigmoid':
            return self._sigmoid
        elif activation_function == 'tanh':
            return self._tanh
        elif activation_function == 'relu':
            return self._relu
        else:
            raise ValueError("not supported activation function. Choose from 'sigmoid', 'tanh', or 'relu'")
    
    def _sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def _tanh(self, x):
        return np.tanh(x)

    def _relu(self, x):
        return np.maximum(0, x)
```
