```python
# Perceptron implementation with  numpy

class Perceptron:
    def __init__(self, learning_rate=0.01, n_iters=1000, activation_function='step', early_stopping=False):
        self.learning_rate = learning_rate # eta parameter -> controls the strength of the update at each iteration
        self.n_iters = n_iters # number of iterations for training
        self.activation_function = self.set_activation_function(activation_function) # for the perceptron we use the step function
        # self._activation_function(activation_function) # set the activation function based on the provided argument
        self.early_stopping = early_stopping # flag for early stopping
        
    def set_activation_function(self, activation_function):
        if activation_function == 'step':
            self._activation_function = self._step_function
        else:
            raise ValueError("Unsupported activation function. Use 'step'.")

    def _step_function(self, x):
        return np.where(x >= 0, 1, -1) # step function: returns 1 if x >= 0, else returns -1
    # used > 0 because the bias is inside the linear output  w_i * x_i + b, so the decision boundary is at 0
    
    def fit(self, X, y, verbose=False):
        """
        Args:
            X: training data, shape (n_samples, n_features)
            y: target labels, shape (n_samples,)

        Fits the model to the training data and updates the weights and bias using the perceptron learning rule
        That is 
            - For each training sample, we compute the linear output (weighted sum of inputs + bias)
            - We apply the activation function to get the predicted label
            - We update the weights and bias based on the error (difference between predicted and true label)
        """
        training_error = [] # to store the error at each iteration for visualization
        nsamples , nFeatures = X.shape

        self.w = np.zeros(nFeatures) # initialize weights to zero
        self.b = 0 # initialize bias to zero

        for _ in range(self.n_iters):
            for idx, x_i in enumerate(X):
                linear_output = np.dot(x_i, self.w) + self.b # compute the aggregate input (weighted sum of inputs + bias)
                y_predicted = self._activation_function(linear_output) # apply the activation function to get the predicted label

                error = y[idx] - y_predicted # compute the error (difference between predicted and true label)
                # perceptron learning rule
                # w = w + learning_rate * error * x_i
                self.w = self.w + self.learning_rate * error * x_i # update weights based on the error and input features
                self.b = self.b + self.learning_rate * error # update bias based on the error
            if self.early_stopping and error == 0: # if early stopping is enabled and there is no error, stop training
                break
            training_error.append(error) # store the error for visualization
            if verbose:
                print(f"Iteration {_+1}/{self.n_iters}, Weights: {self.w}, Bias: {self.b}, Error: {error}")

        self.w = self.w # store the final weights
        self.b = self.b # store the final bias
        self.training_error = training_error # store the training error for visualization

    def predict(self, X):
        if not hasattr(self, 'w') or not hasattr(self, 'b'):
            raise ValueError("Model has not been trained yet. Call fit() before predict().")
        linear_output = np.dot(X, self.w) + self.b # compute the aggregate input (weighted sum of inputs + bias)
        y_predicted = self._activation_function(linear_output) # apply the activation function to get the predicted labels
        return y_predicted
```

```python
# testing with some simple data
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]]) # input features
y = np.array([-1, -1, -1, 1]) # target labels (AND logic gate)

clf = Perceptron(learning_rate=0.1, n_iters=10)
clf.fit(X, y, verbose=True)

predictions = clf.predict(X)
print("Predictions:", predictions)
```
